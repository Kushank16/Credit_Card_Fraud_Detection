```python
import pandas as pd
import numpy as np
```

```python
battle=pd.read_csv("/content/2505213-battles.csv")
```

```python
battle.head()
```

```python
battle.shape  #Its Tells about No. of rows & Column
```

```python
battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True) #here We rename the some names..
battle.head()
```

```python
battle.rename(columns={'defender_1':'primary_defender'},inplace=True) #here We rename the some names..
battle.head()
```

```python
battle['attacker_king'].value_counts() #how many times kings started attacks..
```

```python
battle['location'].value_counts() #Where the battle will occured..
```

```python
#Libraies for data visualization
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.show()
```

```python
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
plt.show()
```
